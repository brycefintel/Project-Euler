```python
def is_prime(number):
    if number==0:
        return "invalid input 0"
    if number==1:
        return True
    for x in range(2,number):
        if number%x==0:
            return False
    return True
```

```python
for x in primes.copy():
    if x > 1000:
        primes.remove(x)
```

```python
primes
```

```python
def check_primes(a,b):
    coeffs=(a*b)
    n=0
    while True:
        if is_prime((n**2)+(a*n)+b):
            n+=1
        else:
            return (a,b,coeffs,n) 
        
```

```python
results={}
for b in primes:
    for a in range(-1000,1000):
        a,b,key,value=check_primes(a,b)
        results[key]=value,a,b
        
```

```python
sorted(results.items(),key=lambda x:x[1],reverse=True)
```
